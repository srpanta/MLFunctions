```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotf(x,y,funcname,xmin,xmax,ymin,ymax,lwidth):
    plt.figure(figsize=(8,6))
    plt.plot(x,y,"r--")
    plt.hlines((ymax+ymin)/2,xmin,xmax,"k")
    plt.vlines((xmax+xmin)/2,ymin,ymax,"k")
    plt.xlim(xmin,xmax) 
    plt.ylim(ymin,ymax) 
    plt.xlabel("x",fontsize=16) 
    plt.ylabel(funcname.split()[0]+'(x)',fontsize=16) 
    plt.title(funcname,fontsize=20) 
    plt.grid()
    plt.show()
```

## Sigmoid Function

```python
x=np.arange(-10,10,0.001)
den = 1+np.exp(-x) 
y=1/den
```

```python
plotf(x,y,"Sigmoid Function",-10,10,-0.1,1.1,3)
```

## Step Function

```python
x=np.arange(-10,10,0.001)
def stepf(x):
    if x>0:
        return 1 
    else: 
        return 0
y=[stepf(i) for i in x]
```

```python
plotf(x,y,"Step Function",-10,10,-0.1,1.1,3)
```

## Softmax Function

```python
x=np.array([1.3, 5.1, 2.2, 0.7, 1.1])
den=np.sum(np.exp(x))
y=np.exp(x)/den
plt.figure(figsize=(8,6))
plt.plot(x,y,"ro")
plt.hlines(0.5,-10,10,"k")
plt.vlines(0,-0.1,1.1,"k")
plt.xlim(-10,10) 
plt.ylim(-0.1,1.1) 
plt.xlabel("x") 
plt.ylabel("S(x)") 
plt.title("Softmax Function",fontsize=20) 
plt.grid()
plt.show()
```

## ReLu Function

```python
def ReLu(x):
    if x>0:
        return x
    else:
        return 0
```

```python
x=np.arange(-10,10,1)
y =[ReLu(i) for i in x]
```

```python
plotf(x,y,"ReLu Function",-10,10,-10,10,3)
```

## Leaky ReLu function

```python
def lReLu(x,a):
    if x>0:
        return x
    else:
        return a*x
```

```python
xmin=-5
xmax=-xmin
x=np.arange(xmin,xmax,0.01)
a=0.1
y =[lReLu(i,a) for i in x]
plotf(x,y,"Leaky ReLu Function",xmin,xmax,-5,5.1,3)
```

## Tanh Function

```python
xmin=-5
xmax=-xmin
x=np.arange(xmin,xmax,0.01)
y =np.tanh(x)
plotf(x,y,"Tanh Function",xmin,xmax,-1.1,1.1,3)
```
